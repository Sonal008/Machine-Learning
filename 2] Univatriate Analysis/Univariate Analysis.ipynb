# 1) Load the dependencies

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.formula.api as smf

import warnings
warnings.filterwarnings("ignore")
```

# 2) Load the dataset 

```python
news = pd.read_csv("NewspaperData.csv")
news.head()
```

```python
news.tail()
```

# 3) Univariate Analysis

- ### 4.1 kde     - (Kernel Distribution Estimation Plot)
- ### 4.2 distplot- (Distribution plot)
- ### 4.3 histplot- (istogram plot)
- ### 4.4 boxplot

```python
# How is data distributed on curve?

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# daily column
ax1.set_title('Daily')
sns.kdeplot(news['daily'], ax=ax1)

# sunday column
ax2.set_title('Sunday')
sns.kdeplot(news['sunday'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# daily column
ax1.set_title('Daily')
print(f'Skewness in daily: {news.daily.skew()}')
print('---------------------------------------')
sns.distplot(news['daily'], ax=ax1)

# sunday column
ax2.set_title('Daily')
sns.distplot(news['daily'], ax=ax2)
print(f'Skewness in sunday: {news.daily.skew()}')
print('---------------------------------------')
plt.show()
```

```python
# plotting first histogram
plt.hist(news['daily'], label='Daily', alpha=.7)
 
# plotting second histogram
plt.hist(news['sunday'], label='Sunday', alpha=0.6)
plt.legend()
 
# Showing the plot using plt.show()
plt.show()
```

```python
# create grouped boxplot 
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# daily column
ax1.set_title('Daily')
sns.boxplot(x = news['daily'], ax=ax1)

# sunday column
ax2.set_title('Sunday')
sns.boxplot(x = news['sunday'], ax=ax2)
plt.show()
```
